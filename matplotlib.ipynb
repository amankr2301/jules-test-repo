# Matplotlib WorkFlow Example 

```python
# import matplotlib and get it ready to be used in jupyter 
%matplotlib inline 
import matplotlib.pyplot as plt 

# Prepare data 
x = [1,2,3,4]
y = [11,22,33,44]

# setup plot 

fig , axes = plt.subplots(figsize=(10,10))  #width and height 

axes.plot(x,y)

# Customize your plot 

axes.set(title="Simple Plot" , 
        xlabel="X-Axis" , 
        ylabel = "Y-Axis")


# save and show the fig(you save the whole figure)

fig.savefig("/kaggle/working/sample.png")


```

<!-- Line Plot -->

```python
# Line plot
import numpy as np 

# linspace Return evenly spaced numbers over a specified interval.(start , stop ,number of samples to be generated )

x = np.linspace(0,10,100)
fig , ax = plt.subplots(figsize=(10,5))
ax.plot(x , x**2)

```

```python
%matplotlib inline 
import matplotlib.pyplot as plt 

x = np.linspace(0, 2 * np.pi, 100)

y = np.sin(x)




fig , ax = plt.subplots(figsize=(10,5))
ax.plot(x,y)

```

```python
# Scatter plot 

# x = np.linspace(0, 10, 100) 

fig , ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# Making a plot from dictionary 
# Making a bar plot 

nut_butter_prices = {
    "Almond_butter" : 10 , 
    "Peanut_butter" : 8 , 
    "Cashew_butter" : 15
}

fig , ax = plt.subplots()

ax.bar(nut_butter_prices.keys() , nut_butter_prices.values())

ax.set(title="Nut Butter Prices" ,
      xlabel="Butter" , 
      ylabel="Prices") ;
```
