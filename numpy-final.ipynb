# Creating Arrays 

```python
ones = np.ones(5)
ones 

one = np.ones((2,5))
one
```

```python
# start from 0 stop at 10 with a jump of 2 
range_array = np.arange(0,10,2)
range_array 
```

```python
# this is just going to have a number between 0 to 10 but if you want an array 
# you need to specify the size like the example below
random = np.random.randint(0,10)
random 

# creating a random array of size 3 * 5 

random_array = np.random.randint(0,10,size=(3,5))
random_array 
```

```python
# This gives a random array of size specified i.e 3x5 having numbers between 0 to 1 
# Note that it is not randint but random 
random_array2 = np.random.random((3,5))
random_array2
```

# Random Seed 


```python
# It is Used to store your random number in case you want to share your notebook so 
# you dont get different random numbers everytime you run the cell 

# The number inside it can be anything its just that whenver it sees 4 it returns that 
# particular random array 
np.random.seed(4)
random_array5 = np.random.randint(10, size=(3,5))
random_array5
```

# Viewing Arrays and Matrices
