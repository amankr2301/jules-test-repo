# Reshaping and transpose 

```python
a2 = np.random.random((2,3))
a2
```

```python
a3 = np.random.random((2,3,3))
a3
```

### Here a2 * a3 is not possible we would need to reshape it 


we have reshaped a2 into 2,3,1 as rule of broadcasting says either the shape
is same or one of them is one. 


```python
a2_reshape = a2.reshape(2,3,1)
# now a2 * a3 is possible 
a2_reshape* a3
```

# Dot Product example (sales_data)



```python
sales_amounts = np.random.randint(20 , size=(5,3))
sales_amounts
```

```python
import pandas as pd 
weekly_sales = pd.DataFrame(sales_amounts , index=["Mon","Tue","Wed","Thu","Fri"],
                           columns = ["Almond" , "Peanut" , "Cashew"])

weekly_sales

```

# Creating prices section 

```python

prices = np.array([10,12,15])


# Here notice that shape of prices is 3, but to put in dataFrame you want 1x3 so 
# reshape it 

butter_prices = pd.DataFrame(prices.reshape(1,3) , index=["Price"] , columns = ["Almond" , 
                                                                  "Peanut",
                                                                  "Cashew"])
butter_prices

```

```python
butter_prices
```

```python
# daily_sales = np.dot(weekly_sales,butter_prices.T)
# Total_price = pd.DataFrame(daily_sales ,index=["Mon","Tue","Wed","Thu","Fri"],
                          # columns = ["Total_Sale"])

# the above dot product gives ndarray and the below gives dataframe 
daily_sales = weekly_sales.dot(butter_prices.T)


daily_sales 


```

```python
weekly_sales["Total"] = daily_sales
```

```python
weekly_sales
```
