```python
import pandas as pd 
import numpy as np 
%matplotlib inline 
import matplotlib.pyplot as plt 
```

```python
heart_disease = pd.read_csv("/kaggle/input/heart-disease/heart-disease.csv")
```

```python
# creating x (features matrix that is everything except target)

x = heart_disease.drop("target" , axis = 1)

# creating y (labels) 

y = heart_disease["target"]
```

```python
# choose the rigth model and hyperparameters

from sklearn.ensemble import RandomForestClassifier #randomforestclassifier is a class ification machine learning model , its capable of learning patters in data and classifying 
model = RandomForestClassifier()

# we will keep the default parameters by the below command we can check what paramters 
# our model is using 

model.get_params()


```

```python
# fit the model to the training data 

from sklearn.model_selection import train_test_split  

# test size is 0.2 that means 20% data hence 80% will be used for training 


x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2)
```

```python
# its telling the classifier model to find the patterns in the training data 

model.fit(x_train , y_train) ; 
```

```python
# making a prediction on the test data y_preds is a conventional name for that 
# but not compulsory 

y_preds = model.predict(x_test)
# y_preds 

```

```python
# Evaluate the model on the training data 
model.score(x_train , y_train) # 1.0 is 100 percent 
```

```python
# Evaluate the model on the test data 
model.score(x_test , y_test) 
```
